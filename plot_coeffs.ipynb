```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import os
import matplotlib.colors as mcolors
from scipy.stats import ttest_ind

# === Settings ===
base_path = "/Users/Joey/Volumes/orion/JediCode/hybrid_tlm/jedi-bundle/build/fv3-jedi/test/Data"
npz_levels = [3, 5, 7, 9]
modes = [1, 2, 3]
vector_indices = range(6)
level_index = 63
var_name = "eastward_wind"

# === Function to load and reshape data ===
def load_data(npz, mode, vector_index):
    dir_name = f"hyb_40_ens_membs_{npz}_lvls"
    filename = f"hybrid_linear_model_coeffs_20211221T000000Z_000006-00000{mode}.nc"
    full_path = os.path.join(base_path, dir_name, filename)
    try:
        ds = xr.open_dataset(full_path)
        data_1d = ds[var_name][:, level_index, vector_index].values
        return data_1d.reshape((48, 48))
    except FileNotFoundError:
        print(f"Missing: {full_path}")
        return None

# === Create custom colormap: dark red → white, mask as gray ===
colors = [(0.4, 0, 0), (1, 1, 1)]  # deep red to white
custom_cmap = mcolors.LinearSegmentedColormap.from_list("sig_cmap", colors)
boundaries = np.linspace(0, 0.05, 256)
norm = mcolors.BoundaryNorm(boundaries, ncolors=256)

# === Set up plot ===
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(18, 9))
axes = axes.flatten()

row_idx = 0
for mode in modes:
    for vector_index in vector_indices:
        ax = axes[row_idx]
        
        # Load all 3 fields
        data_fields = {}
        for npz in npz_levels:
            data = load_data(npz, mode, vector_index)
            if data is not None:
                data_fields[npz] = data

        if len(data_fields) < 4:
            ax.axis('off')
            row_idx += 1
            continue

        # Compute p-values for upper triangle
        pval_matrix = np.full((4, 4), np.nan)
        for i, npz1 in enumerate(npz_levels):
            for j, npz2 in enumerate(npz_levels):
                if i < j:
                    data1 = data_fields[npz1].flatten()
                    data2 = data_fields[npz2].flatten()
                    tstat, pval = ttest_ind(data1, data2, equal_var=False)
                    pval_matrix[i, j] = pval

        # Mask diagonal and lower triangle
        mask = np.tril(np.ones_like(pval_matrix, dtype=bool))  # True where we want to gray it out
        masked_pvals = np.ma.masked_array(pval_matrix, mask=mask)

        # Plot masked array
        cmap_with_bad = custom_cmap.copy()
        cmap_with_bad.set_bad(color='lightgray')  # gray for masked values

        cax = ax.matshow(masked_pvals, cmap=cmap_with_bad, norm=norm)
        ax.set_xticks(range(4))
        ax.set_yticks(range(4))
        ax.set_xticklabels(npz_levels, fontsize=8)
        ax.set_yticklabels(npz_levels, fontsize=8)
        ax.set_title(f"Mode {mode}, Vec {vector_index}", fontsize=10)
        row_idx += 1

# Hide any unused axes
for i in range(row_idx, len(axes)):
    axes[i].axis('off')

# Add colorbar
fig.subplots_adjust(right=0.88, wspace=0.4, hspace=0.6)
cbar_ax = fig.add_axes([0.90, 0.2, 0.02, 0.6])
cb = fig.colorbar(cax, cax=cbar_ax)
cb.set_label("Significance Level (p < 0.05)", fontsize=12, weight='bold')
cb.ax.tick_params(labelsize=10)

# Title
fig.suptitle("Statistically Significant Differences Between Vertical Averaging Levels", fontsize=16, y=0.98, weight='bold')
plt.show()
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import os
import matplotlib.colors as mcolors
from scipy.stats import ttest_ind, mannwhitneyu, shapiro

# === Settings ===
base_path = "/Users/Joey/Volumes/orion/JediCode/hybrid_tlm/jedi-bundle/build/fv3-jedi/test/Data"
npz_levels = [3, 5, 7, 9]
modes = [1, 2, 3]
vector_indices = range(6)
level_index = 63  # can be ignored when averaging over levels later
var_name = "eastward_wind"

# === Function to load and vertically average data ===
def load_data(npz, mode, vector_index):
    dir_name = f"hyb_40_ens_membs_{npz}_lvls"
    filename = f"hybrid_linear_model_coeffs_20211221T000000Z_000006-00000{mode}.nc"
    full_path = os.path.join(base_path, dir_name, filename)
    try:
        ds = xr.open_dataset(full_path)
        # Average over vertical dimension
        data = ds[var_name][:, :, vector_index].mean(dim='nz_eastward_wind').values  # shape (2304,)
        return data.reshape((48, 48))
    except FileNotFoundError:
        print(f"Missing: {full_path}")
        return None

# === Custom colormap: dark red → white; non-sig or masked as gray ===
colors = [(0.4, 0, 0), (1, 1, 1)]  # deep red to white
custom_cmap = mcolors.LinearSegmentedColormap.from_list("sig_cmap", colors)
boundaries = np.linspace(0, 0.05, 256)
norm = mcolors.BoundaryNorm(boundaries, ncolors=256)

# === Plot setup ===
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(18, 9))
axes = axes.flatten()
row_idx = 0

for mode in modes:
    for vector_index in vector_indices:
        ax = axes[row_idx]

        # Load and store data for this mode/vector combo
        data_fields = {}
        for npz in npz_levels:
            data = load_data(npz, mode, vector_index)
            if data is not None:
                data_fields[npz] = data

        # Skip if any level missing
        if len(data_fields) < 4:
            ax.axis('off')
            row_idx += 1
            continue

        # Initialize p-value matrix
        pval_matrix = np.full((4, 4), np.nan)
        for i, npz1 in enumerate(npz_levels):
            for j, npz2 in enumerate(npz_levels):
                if i < j:
                    data1 = data_fields[npz1].flatten()
                    data2 = data_fields[npz2].flatten()

                    # Use a subsample for Shapiro-Wilk to keep things fast
                    sample1 = data1[:5000] if len(data1) > 5000 else data1
                    sample2 = data2[:5000] if len(data2) > 5000 else data2
                    try:
                        _, p_shapiro1 = shapiro(sample1)
                        _, p_shapiro2 = shapiro(sample2)
                    except Exception:
                        # fallback in case Shapiro fails (e.g., constant arrays)
                        p_shapiro1, p_shapiro2 = 0, 0

                    if p_shapiro1 > 0.05 and p_shapiro2 > 0.05:
                        # Gaussian: use t-test
                        _, pval = ttest_ind(data1, data2, equal_var=False)
                    else:
                        # Non-Gaussian: use Mann–Whitney U test
                        try:
                            _, pval = mannwhitneyu(data1, data2, alternative='two-sided')
                        except ValueError:
                            pval = np.nan  # fallback if identical distributions

                    pval_matrix[i, j] = pval

        # Mask lower triangle & diagonal
        mask = np.tril(np.ones_like(pval_matrix, dtype=bool))
        masked_pvals = np.ma.masked_array(pval_matrix, mask=mask)

        # Plot significance matrix
        cmap_with_bad = custom_cmap.copy()
        cmap_with_bad.set_bad(color='lightgray')

        cax = ax.matshow(masked_pvals, cmap=cmap_with_bad, norm=norm)
        ax.set_xticks(range(4))
        ax.set_yticks(range(4))
        ax.set_xticklabels(npz_levels, fontsize=8)
        ax.set_yticklabels(npz_levels, fontsize=8)
        ax.set_title(f"Mode {mode}, Vec {vector_index}", fontsize=10)
        row_idx += 1

# Hide unused subplots
for i in range(row_idx, len(axes)):
    axes[i].axis('off')

# Add colorbar
fig.subplots_adjust(right=0.88, wspace=0.4, hspace=0.6)
cbar_ax = fig.add_axes([0.90, 0.2, 0.02, 0.6])
cb = fig.colorbar(cax, cax=cbar_ax)
cb.set_label("Significance Level (p)", fontsize=12, weight='bold')
cb.ax.tick_params(labelsize=10)

# Title
fig.suptitle("Significance of Vertical Averaging Differences (Gaussianity-Adaptive)", fontsize=16, y=0.98, weight='bold')
plt.show()
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os

# === Settings ===
base_path = "/Users/Joey/Volumes/orion/JediCode/hybrid_tlm/jedi-bundle/build/fv3-jedi/test/Data"
npz_levels = [3, 5, 7, 9]
modes = [0, 1, 2]  # Mode indices (0-based)
tiles = range(1, 7)  # Cube face numbers (1–6)
level_index = 63
var_name = "eastward_wind"

for tile in tiles:
    fig = plt.figure(figsize=(4 * len(npz_levels) + 1, 3 * len(modes)))
    gs = gridspec.GridSpec(len(modes), len(npz_levels) + 1, width_ratios=[1]*len(npz_levels) + [0.05], wspace=0.2)
    row_vmax = []

    # === Precompute vmax per mode for consistent colorbar ===
    for mode in modes:
        vmax_list = []
        for npz in npz_levels:
            dir_name = f"hyb_40_ens_membs_{npz}_lvls"
            filename = f"hybrid_linear_model_coeffs_20211221T000000Z_000006-00000{tile:01d}.nc"
            full_path = os.path.join(base_path, dir_name, filename)

            try:
                ds = xr.open_dataset(full_path)
                data_1d = ds[var_name][:, level_index, mode].values
                data_2d = data_1d.reshape((48, 48))
                vmax_list.append(np.max(np.abs(data_2d)))
            except (FileNotFoundError, IndexError):
                continue

        row_vmax.append(max(vmax_list) if vmax_list else 1.0)

    # === Plot panels for this tile ===
    for i, mode in enumerate(modes):
        vmax = row_vmax[i]
        im_last = None

        for j, npz in enumerate(npz_levels):
            ax = fig.add_subplot(gs[i, j])
            dir_name = f"hyb_40_ens_membs_{npz}_lvls"
            filename = f"hybrid_linear_model_coeffs_20211221T000000Z_000006-00000{tile:01d}.nc"
            full_path = os.path.join(base_path, dir_name, filename)

            try:
                ds = xr.open_dataset(full_path)
                data_1d = ds[var_name][:, level_index, mode].values
                data_2d = data_1d.reshape((48, 48))

                im = ax.pcolormesh(data_2d, shading='auto', cmap='RdBu_r', vmin=-vmax, vmax=vmax)
                if j == 0:
                    ax.set_ylabel(f"Mode {mode + 1}", fontsize=10)
                if i == 0:
                    ax.set_title(f"npz = {npz}")
                im_last = im

            except (FileNotFoundError, IndexError):
                ax.set_title("Missing")
                ax.axis('off')
                print(f"Missing or error: {full_path}")

        if im_last is not None:
            cax = fig.add_subplot(gs[i, -1])
            fig.colorbar(im_last, cax=cax)

    fig.suptitle(f"{var_name} at level {level_index} — Tile {tile}", fontsize=16)
    plt.tight_layout(rect=[0, 0, 0.95, 0.96])
    plt.show()
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import os
import matplotlib.colors as mcolors
from scipy.stats import ttest_ind, mannwhitneyu, shapiro

# === Settings ===
base_path = "/Users/Joey/Volumes/orion/JediCode/hybrid_tlm/jedi-bundle/build/fv3-jedi/test/Data"
npz_levels = [3, 5, 7, 9]
modes = [0, 1, 2]  # Mode indices (0-based!)
tiles = range(1, 7)  # Cube faces
var_name = "eastward_wind"

# === Function to load and vertically average data ===
def load_data(npz, mode, tile):
    dir_name = f"hyb_40_ens_membs_{npz}_lvls"
    filename = f"hybrid_linear_model_coeffs_20211221T000000Z_000006-00000{tile:01d}.nc"
    full_path = os.path.join(base_path, dir_name, filename)
    try:
        ds = xr.open_dataset(full_path)
        data = ds[var_name][:, :, mode].mean(dim='nz_eastward_wind').values  # shape (2304,)
        return np.abs(data.reshape((48, 48)))  # ⬅️ Take absolute value here
    except FileNotFoundError:
        print(f"Missing: {full_path}")
        return None

# === Custom colormap: deep red to white; masked = gray ===
colors = [(0.4, 0, 0), (1, 1, 1)]  # dark red → white
custom_cmap = mcolors.LinearSegmentedColormap.from_list("sig_cmap", colors)
boundaries = np.linspace(0, 0.05, 256)
norm = mcolors.BoundaryNorm(boundaries, ncolors=256)

# === Plot setup ===
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(18, 9))
axes = axes.flatten()
row_idx = 0

for mode in modes:
    for tile in tiles:
        ax = axes[row_idx]

        # Load and store vertically averaged fields
        data_fields = {}
        for npz in npz_levels:
            data = load_data(npz, mode, tile)
            if data is not None:
                data_fields[npz] = data

        if len(data_fields) < 4:
            ax.axis('off')
            row_idx += 1
            continue

        # Compute significance matrix
        pval_matrix = np.full((4, 4), np.nan)
        for i, npz1 in enumerate(npz_levels):
            for j, npz2 in enumerate(npz_levels):
                if i < j:
                    data1 = data_fields[npz1].flatten()
                    data2 = data_fields[npz2].flatten()

                    # Subsample for Shapiro
                    sample1 = data1[:5000] if len(data1) > 5000 else data1
                    sample2 = data2[:5000] if len(data2) > 5000 else data2
                    try:
                        _, p_shapiro1 = shapiro(sample1)
                        _, p_shapiro2 = shapiro(sample2)
                    except Exception:
                        p_shapiro1, p_shapiro2 = 0, 0

                    # Gaussian or not?
                    if p_shapiro1 > 0.05 and p_shapiro2 > 0.05:
                        _, pval = ttest_ind(data1, data2, equal_var=False)
                    else:
                        try:
                            _, pval = mannwhitneyu(data1, data2, alternative='two-sided')
                        except ValueError:
                            pval = np.nan
                    pval_matrix[i, j] = pval

        # Mask lower triangle + diagonal
        mask = np.tril(np.ones_like(pval_matrix, dtype=bool))
        masked_pvals = np.ma.masked_array(pval_matrix, mask=mask)

        # Plot matrix
        cmap_with_bad = custom_cmap.copy()
        cmap_with_bad.set_bad(color='lightgray')
        cax = ax.matshow(masked_pvals, cmap=cmap_with_bad, norm=norm)
        ax.set_xticks(range(4))
        ax.set_yticks(range(4))
        ax.set_xticklabels(npz_levels, fontsize=8)
        ax.set_yticklabels(npz_levels, fontsize=8)
        ax.set_title(f"Mode {mode+1}, Tile {tile}", fontsize=10)
        row_idx += 1

# Hide unused subplots
for i in range(row_idx, len(axes)):
    axes[i].axis('off')

# Colorbar and formatting
fig.subplots_adjust(right=0.88, wspace=0.4, hspace=0.6)
cbar_ax = fig.add_axes([0.90, 0.2, 0.02, 0.6])
cb = fig.colorbar(cax, cax=cbar_ax)
cb.set_label("Significance Level (p)", fontsize=12, weight='bold')
cb.ax.tick_params(labelsize=10)

# Title
fig.suptitle("Significance of Magnitude Differences (Sign-agnostic, Gaussianity-Adaptive)", fontsize=16, y=0.98, weight='bold')
plt.show()
```
